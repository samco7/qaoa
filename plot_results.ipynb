```python
import pickle
from save_results import *
from matplotlib import pyplot as plt
plt.style.use('seaborn')
```

```python
# load the saved dictionary holding the results
res = load_result('./results/2022-07-19/2022-07-19_results_1.pickle')

# initialize lists to hold results
points = []
layers = []
accuracy_20_layers = []
accuracy_12_points = []
layers_v_accuracy = []
runtime_20_points = []
runtime_20_layers = []
simtime_20_points = []
simtime_20_layers = []

# parse results out of loaded dictionary
for item in res:
    if item[0][0:2] == '20':
        accuracy_20_layers.append(res[item]['approx_ratio'])
        points.append(int(item[1][0:2]))
        runtime_20_layers.append(res[item]['device_time'])
        simtime_20_layers.append(res[item]['sim_time'])
    if item[1][0:2] == '12':
        accuracy_12_points.append(res[item]['approx_ratio'])
        layers.append(int(item[0][0:2]))
    if item[1][0:2] == '20':
        runtime_20_points.append(res[item]['device_time'])
        simtime_20_points.append(res[item]['sim_time'])
```

```python
# plot number of points vs sim runtime for 20 layers
plt.figure(figsize=(12, 8))
plt.plot(points, simtime_20_layers)
plt.title('Number of Points vs Simulation Runtime for 20 Layers', fontsize=18)
plt.xlabel('Number of Points', fontsize=14)
plt.ylabel('Average Simulation Runtime', fontsize=14)
plt.show()
```

```python
# plot number of points vs runtime for 20 layers
plt.figure(figsize=(12, 8))
plt.plot(points, runtime_20_layers)
plt.title('Number of Points vs Estimated Device Time for 20 Layers', fontsize=18)
plt.xlabel('Number of Points', fontsize=14)
plt.ylabel('Average Device Time Estimate', fontsize=14)
plt.show()
```

```python
# plot number of points vs accuracy for 20 layers
plt.figure(figsize=(12, 8))
plt.plot(points, accuracy_20_layers)
plt.title('Number of Points vs Accuracy for 20 Layers', fontsize=18)
plt.xlabel('Number of Points', fontsize=14)
plt.ylabel('Average Accuracy', fontsize=14)
plt.show()
```

```python
# plot number of points vs sim runtime for 20 points
plt.figure(figsize=(12, 8))
plt.plot(layers, simtime_20_points)
plt.title('Number of Layers vs Simulation Runtime for 20 Points', fontsize=18)
plt.xlabel('Number of Points', fontsize=14)
plt.ylabel('Average Simulation Runtime', fontsize=14)
plt.show()
```

```python
# plot number of layers vs runtime for 20 points
plt.figure(figsize=(12, 8))
plt.plot(layers, runtime_20_points)
plt.title('Number of Layers vs Estimated Device Time for 20 Points', fontsize=18)
plt.xlabel('Number of Layers', fontsize=14)
plt.ylabel('Average Device Time Estimate', fontsize=14)
plt.show()
```
