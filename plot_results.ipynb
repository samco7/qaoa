```python
import pickle
from save_results import *
from matplotlib import pyplot as plt
plt.style.use('seaborn')
```

```python
# load the saved dictionary holding the results
res = load_result('./results/2022-07-21/2022-07-21_results_1.pickle')

# initialize lists to hold results
n_layers = 20 #held constant
n_points = 15 #held constant
points = []
layers = []
accuracy_n_layers = []
accuracy_n_points = []
layers_v_accuracy = []
runtime_n_points = []
runtime_n_layers = []
simtime_n_points = []
simtime_n_layers = []

# parse results out of loaded dictionary
for key in res: #each key is a tuple of the form (n_layers, n_points)
    if key[0] == n_layers:
        accuracy_n_layers.append(res[key]['approx_ratio'])
        points.append(key[1])
        runtime_n_layers.append(res[key]['device_time'])
        simtime_n_layers.append(res[key]['sim_time'])
    if key[1] == n_points:
        accuracy_n_points.append(res[key]['approx_ratio'])
        layers.append(key[0])
        runtime_n_points.append(res[key]['device_time'])
        simtime_n_points.append(res[key]['sim_time'])
```
